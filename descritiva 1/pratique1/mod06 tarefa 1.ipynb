#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
base = pd.read_csv('sinasc_RO_2019.csv')
base.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.rc('figure', figsize=(12,8))
fig, axes = plt.subplots(2,3)
fig.delaxes(ax=axes[1,2])

sns.countplot(ax=axes[0,0], x='APGAR1', data=base)
sns.countplot(ax=axes[0,1], x='APGAR5', data=base)
sns.countplot(ax=axes[0,2], x='SEXO', data=base)
sns.countplot(ax=axes[1,0], x='GESTACAO', data=base).tick_params(axis='x', rotation=30)
sns.countplot(ax=axes[1,1], x='ESTCIVMAE', data=base)
plt.subplots_adjust(right=11, left=9.9)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
plt.figure('figure', figsize=(10,8))
fig, axes = plt.subplots(1, 2)

sns.countplot(ax=axes[0], x='APGAR5', data=base)
sns.barplot(ax=axes[1], x='APGAR5', y='CONSULTAS', data=base)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
fig, axes = plt.subplots(2,2)
sns.countplot(ax=axes[0,0], x='SEXO', data=base)
sns.countplot(ax=axes[0,1], x='GESTACAO', data=base).tick_params(axis='x', rotation=20)
sns.countplot(ax=axes[1,0], x='ESTCIVMAE', data=base)
sns.countplot(ax=axes[1,1], x='CONSULTAS', data=base)
plt.subplots_adjust(top=11, bottom=9.9)
```
